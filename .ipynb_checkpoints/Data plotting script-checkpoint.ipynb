```python
import numpy as np, scipy.stats as st
import matplotlib.pyplot as plt
```

```python

mean = lambda a: np.mean(a)
ci = lambda a: st.t.interval(0.95, len(a)-1, loc=np.mean(a), scale=st.sem(a))
```

```python
files = {}
data = {}
data_keywords = ["success_rate", "average_reward", "average_initial_value"]
method_keywords = ["ratio", "two_goal", "one_goal"]
base_loc = "/home/liam/Desktop/usher_repo/USHER/psst/logging/gridworld_results/"
for keyword in method_keywords:
    with open(base_loc + keyword + ".txt", "r") as f: 
        files[keyword] = f.read().split("\n")
        
    data[keyword] = {}
    for data_key in data_keywords:
        data[keyword][data_key] = [float(s.split(":")[-1]) for s in files[keyword] if data_key in s]
```

```python
means = {k1 : {k2 : mean(data[k1][k2]) for k2 in data[k1].keys()} for k1 in data.keys()}
cis = {k1 : {k2 : ci(data[k1][k2]) for k2 in data[k1].keys()} for k1 in data.keys()}
std_err = {k1 : {k2 : (cis[k1][k2][1]-cis[k1][k2][0])/2 for k2 in data[k1].keys()} for k1 in data.keys()}
```

```python
# means
# cis
std_err
```

```python
x_pos = np.arange(len(method_keywords))
def extract(dictionary, measure):
    return [dictionary[k][measure] for k in method_keywords]
for measure in data_keywords: 
    fig, ax = plt.subplots()
    ax.bar(x_pos, extract(means, measure), yerr=extract(std_err, measure), alpha=0.5, ecolor='black', capsize=10)
    ax.set_ylabel(measure)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(method_keywords)
    ax.set_title(f"Comparison of {measure} across each method")
#     ax.yaxis.grid(True)
    plt.tight_layout()
    plt.savefig(f"{base_loc}performance on navigation env -- {measure}.png" )
    plt.show()
```
